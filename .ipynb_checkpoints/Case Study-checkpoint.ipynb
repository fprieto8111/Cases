```python
def lower_case_column_names(y):
    y.columns=[i.lower() for i in y.columns]
    return y

lower_case_column_names(NDF)
```

```python
def drop_columns(x) :
    x.drop(columns=["customer"], inplace=True)
    return x

drop_columns(NDF)
```

```python
def clean_gender(z):
    if z in ['M',"Male"]:
        return 'Male'
    elif z in ['F', 'female',"Femal"]:
        return 'Female'
    elif np.nan:  
        pass
    else:
        return 'U'
    
NDF['gender'] = NDF['gender'].apply(clean_gender)
NDF
```

```python
def clean_state(state):
    if state=="Cali":
        return "California"
    elif state=="AZ":
        return "Arizona"
    elif state=="WA":
        return "Washington"
    else:
        return state
    
NDF['state'] = NDF['state'].apply(clean_state)
NDF
```

```python
def clean_edu(z):
    if z in ['Bachelor', 'Bachelors', 'College']:
        return 'Bachelor'
    elif z == 'High School or Below':
        return 'High School or Below'
    elif z == 'Master':
        return 'Master'
    elif z == 'Doctor':
        return 'Doctor'
    elif np.nan:  pass
    else: 
        return z
    
NDF['education'] = list(map(clean_edu,NDF["education"]))
```

```python
NDF.value_counts("education")
```

```python
def cleanint(val):
    if str(val).endswith("%"):
        val = int(round(float(val.rstrip(val[-1]))/100,0))
        return val
    else:
        try:
            return int(round(val))
        except ValueError:
            return val

NDF["customer lifetime value"] = NDF["customer lifetime value"].apply(cleanint)
NDF
```

```python
#replaces NaNs with the mean
mean_val=round(NDF["customer lifetime value"].mean(skipna=True))

NDF["customer lifetime value"]=NDF["customer lifetime value"].fillna(mean_val)
NDF["customer lifetime value"]=NDF["customer lifetime value"].apply(int)

#replaces 0s with mean
mean_life=round(NDF["customer lifetime value"].mean(skipna=True),1)
NDF["customer lifetime value"]=NDF["customer lifetime value"].replace(0.0,mean_life)
```

```python
mean_income=round(NDF["income"].mean(skipna=True),1)
NDF["income"]=NDF["income"].replace(0.0,mean_income)
```

```python
mean_mpa=round(NDF["monthly premium auto"].mean(skipna=True),1)
NDF["monthly premium auto"]=NDF["monthly premium auto"].replace(0.0,mean_mpa)
```

```python
mean_claim=round(NDF["total claim amount"].mean(skipna=True),1)
NDF["total claim amount"]=NDF["total claim amount"].replace(0.0,mean_claim)
```

```python
def cleancomplaints(comp):
    try:
        comp=int(comp)
        return comp
    except ValueError:
        if type(comp) is float:
            return 10101
        else:
            comp=comp.split("/")
            comp=comp[1]
            return int(comp)
        
        
NDF["number of open complaints"] = NDF["number of open complaints"].apply(cleancomplaints)
NDF
```

```python
def bucket_state(state):
    if state=="California":
        return "West Region"
    elif state=="Oregon":
        return "North West Region"
    elif state=="Washington":
        return "East Region"
    elif state=="Arizona" or state=="Nevada":
        return "Central Region"

NDF['state'] = NDF['state'].apply(bucket_state)
NDF
```

```python
def bucket_cars(car):
    if car=="Luxury SUV" or car=="Luxury Car":
        return "Luxury Vehicle"
    else: 
        return car

NDF['vehicle class'] = NDF['vehicle class'].apply(bucket_cars)
NDF
```
